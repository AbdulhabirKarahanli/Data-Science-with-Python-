### Pivot Tables

    Pivot Tables: Multidimensional summary of data
    
    ▪ pivot table takes simple columnwise data as input, and groups the entries into a two-dimensional table that provides a multidimensional summarization of the data
    ▪ like a multidimensional version of GroupBy aggregation (i. e., both the split and the combine happen across a two-dimensional grid)
    
### Pivot Tables: The Titanic Example

```python
titanic = sns.load_dataset('titanic')
titanic.head(3)
```

```python
# we might want to inspect data based on multiple grouping variables
    # → e. g., survival rate not only by gender, but also by class
    
# simple dimentional groupby: 

titanic.groupby('sex')[['survived']].mean()
```

```python
# two dimensional groupby using pivot table

titanic.pivot_table('survived', index='sex', columns='class', aggfunc='mean')
```

### The aggfunc keyword (pivot_table)
    ▪ aggfunc keyword controls what type of aggregation is applied (mean by default)
        → aggregation specification can be a string representing one of several common choices (e. g., sum,
    mean, count, min, max, etc.)
        → also possible: function that implements an aggregation ( np.sum() , min() , sum() , etc.)
    ▪ can be specified as a dictionary mapping a column to any of the desired aggregation specifications

```python
titanic.pivot_table(index='sex', columns='class', aggfunc={'survived':sum, 'fare':'mean'})
```

### The margins keyword (pivot_table)
    ▪ at times it’s useful to compute totals along each grouping: keyword margins

```python
titanic.pivot_table('survived', index='sex', columns='class', margins=True) # default agg func: mean
```

## Text Data

```python
# This is very little part of text data with Python: 
# Regular Python vectorization breaks if data contains missing values. 

data = ['peter', 'Paul', 'MARY', 'gUID']

print(data)
print([s.capitalize() for s in data])  ## the firs letter became capital. 

##The problem is when the dataset has a missing value, the code is not run successfully. 
# we can tackle this problem by using Pandas. 

import pandas as pd 

data1 = ['peter', 'Paul', None, 'MARY', 'gUID']
print(data1)

names = pd.Series(data1)
print(names)

print(names.str.capitalize())  ## It worked out again. 
```

```python
## Example ## 

monte = pd.Series(['Graham Chapman', 'John Cleese', 'Terry Gilliam', 
                   'Eric Idle', 'Terry Jones', 'Micheal Palin']) 

print(monte)

print(monte.str.lower().head())  ##head() fuction itsel works with fist 5 elements by default. 
        # we basically, changed the letter into the lower case. 
        
# so, which names/last names starts with any of the letter? by 'T'? 

print(monte.str.startswith('T')) ## This gives the results in all dataset. 
print(monte.str.startswith('T').head()) #as I said, by default, it is first 5 elements. 
print(monte.str.startswith('T').head(3))  # this gives the first three elemenets. 


print(monte.str.split().head(3))   ## this seperates the names and last names by qotation mark. 
```

```python
### Regular Expressions & Pandas ###
# Mapping between Pandas string mothods and functions in Python's 're' modele
# "re" stands for regular expressions



## Regular Expressions: Example: 

    #we can extract the first name from each series element by asking for a 
    #   contiguous group of characters at the beginning of each elemnet. 
    
print(monte.str.extract('([A-Za-z]+)'))  # From A to Z both upper/lower case, by the plus sign (+)
    #we meant it could be repeated (only these types of characters can be repeated, as soon as any character
    #rahter than this group, like number or space, then this extraction stops). 

# Regular Expression:A more complicated example: 
    
# a question might be: 

#find all names that start and end with a consonant(ünsüz, vowel=sesli/ünlü hafr), making use of the
# start-of-string (^) and end-of-string ($) regular expression characters
#Consonants in English: A, E, I, O, U. Others are vowels. 

# "^" the hat, means inverse/not (for example: [^abd] is used to match a single character not present in the list. 
        #Besides, it means starting point. 

print(monte.str.findall(r'^[^AEIOU].*[^aeiou]$')) #verbal expression of the line is that 
    #find all names that starts with not any of AEIOU (anything could be in between) and 
    #ends without any of aeiou. Because. "a,e,i,o,u" are vowels. Inverse of vowels=consonants
    # As a result, John Cleese is missing because of the 'e' at the end, and Eric Idel as it is clear. 
    
    
## Exercise with persons dataset.
        #How many person's names start with a consonant and end with a vowel?
        
print(persons)
print(persons.name) 

#Answer: 
    
print(persons.name.str.findall(r'^[^AEIOU].*[aeiou]$'))  # I did it at once !!! 


```
