### How to save plots? 
    ▪ create figure object fig = plt.figure() , then use savefig() command to save the plot
        → saves the last created plot, not necessary to use plt.show()
    ▪ wide range of file types supported (e. g., .png, .jpg, .pdf, .svg)
        → method .canvas.get_supported_filetypes() to show all supported file types

```python
import numpy as np

x = np.linspace(0, 10, 100)
```

```python
# Create plot: 

fig = plt.figure()
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x));
```

```python
## Save plot: 

fig.savefig('my_figure.png')
fig.savefig('my_figure1.pdf')
# to show the all supported filetypes:

fig.canvas.get_supported_filetypes()
```

#### There are two types of interfaces for plotting:
    → MATLAB-style state-based interface
    → and a more powerful object-oriented interface (this is more powerful)


#### MATLAB-style interface

```python
import numpy as np; x = np.linspace(0,10,100)


plt.figure() #create a plot figure
# create the first of two panels and set current axis
plt.subplot(2, 1, 1) #(rows, columns, panel number)
plt.plot(x, np.sin(x))
# create the second panel and set the current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

### Object-oriented interface:

```python
import numpy as np; x = np.linspace(0,10,100)

#first create a grid of plots
#ax will be an array of two Axes objects

fig, ax = plt.subplots(2)

#call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```
