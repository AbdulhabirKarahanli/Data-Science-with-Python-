## Simple line plots:
    ▪ .plot(x, y) method to plot vectors of data
        → default is to connect data with a line
        → multiple lines are automatically coded in different colors

```python
import numpy as np 
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
## This is also an example for MATLAB interface

x = np.linspace(1, 10, num=10) #a data vector 
y = np.linspace(1, 10, num=10)
x, y

plt.plot(x,y) # first line (blue)
plt.plot(x,y**2) #second line (y squared)
plt.plot(x, 3*y); #(I tried this - the green line)
```

#### The same example but with object-oriented interface:

```python
# create figure and axes
fig = plt.figure()
ax = plt.axes()

#the figure is ready and now plot the data

ax.plot(x,y)
ax.plot(x, y**2)
ax.plot(x, 3*y);
```

### Adjusting simple plots: 
    1. Changing the line color with the 'color' argument
    2. Changing the line type with the 'linestyle' argument
    3. Adjusting axes limits with .xlim() and .ylim()
    4. Title and axes labels
    5. Data labels & legends

```python
# 1. Color

x = y = np.linspace(1, 10, num=10)

plt.plot(x,y + 0, color='blue')         # specify color by name
plt.plot(x,y + 1, color='g')            # short color code 'rgbcmyk'
plt.plot(x,y + 2, color='0.75')         # gratscale btw 0 and 1
plt.plot(x,y + 3, color='#FFDD44')      # Hex code (RRGGBB from 00 to FF)
plt.plot(x,y + 4, color=(1.0,0.2,0.3))  # RGB tuple, values from 0 and 1
plt.plot(x,y + 5, color='chartreuse');  # all HTML color names supported
```

```python
# 2. Line type: 
plt.plot(x,y + 0, linestyle='solid')
plt.plot(x,y + 1, linestyle='dashed')
plt.plot(x,y + 2, linestyle='dashdot')
plt.plot(x,y + 3, linestyle='dotted')

#also we can use for the same purpose these abrreviations:

plt.plot(x,y + 4, linestyle='-')    #solid
plt.plot(x,y + 5, linestyle='--')   #dashed
plt.plot(x,y + 6, linestyle='-.')   #dashdot
plt.plot(x,y + 7, linestyle=':');    #dotted
```

```python
# 3. Adjusting axes limits with .xlim() and .ylim():

## adjust limits:

plt.xlim(-1, 11);
plt.ylim(-1, 25);
#note: we could also display axes in reverse by reversing the order of the argument(e.g., plt.ylim(20, -1))

plt.plot(x,y)
plt.plot(x, 2*y);
```

```python
# 4. Title and axes lables: 

plt.plot(x,y)
plt.plot(x, 2*y);

#Now add titles: .xlabel("") or .ylabel("") methods
plt.title("A simple line plot")
plt.xlabel("x")
plt.ylabel("f(x)=2*x");
```

```python
# 5. Legends: to specify which line belong to which data we use legends. 
#We just add labels to the respective plots

x = np.linspace(0, 10, 100)

plt.plot(x, np.sin(x), '-g', label='sin(x)') #-g means solid line and green color
plt.plot(x, np.cos(x), ':b', label='cos(x)') #:b means dotted line and blue color
plt.plot(x, np.sin(x+1), '--y')              # legend is missing because of missing label here
plt.axis('equal')                            # equal means I wanna have same or perspective/proportions across bo
plt.legend();                                # show the legend





```

### Visualizing Errors

#### Visualizing errors using errorbars
       ▪ reporting the error around a measurement of great importance
        → two data series can have for example the same mean but completely different underlying data
       ▪ it is good practice to include visualizations of the error
        → plt.errorbar(x, y, yerr) (or xerr) to add errors to a scatterplot (how much big is the error with respect to y is coded as "yerr" and same for x as "xerr".
        

#### Example: Mean order value per account type error

```python
sales = pd.read_csv('/Users/abdulhabirkarahanli/Desktop/Data/sales.csv')
sales

```

```python
import numpy as np
import pandas as pd

# import data

sales = pd.read_csv('/Users/abdulhabirkarahanli/Desktop/Data/sales.csv')
sales

# define a function for standard errors of the mean
# We can have standard deviations with default, but for standard error we should define a function

def se(x):
    """compute standard errors of the mean (needs numpy)"""
    return np.sqrt(x.var() / len(x))  #square root of 'vaiance of x and divided by the length (n)'.

# aggregate data per group 

sales_means = sales.groupby('Account')['Value'].mean()
sales_se = sales.groupby('Account')['Value'].aggregate([se])  #returns dataframe
sales_groups = np.unique(sales['Account'])
df = pd.DataFrame({'group': sales_groups, 'mean': sales_means, 'se': sales_se['se']})

# let me see what we have done: 
sales_means
sales_se
sales_groups
df
# plot results

plt.errorbar(x=df['group'], y=df['mean'], yerr=df['se'], linestyle='None', marker='o')
plt.xlim(-.5, 1.5)
plt.ylabel('Mean order value'); plt.title('A grouped plot with error bars')
plt.show();
```

### Bar plots

#### Bar plot with plt.bar
    ▪ function plt.bar works just like plt.plot and plt.scatter 
        → takes x (numeric or categorical) and y values as input
        → lots of optional arguments like yerr for plotting error bars
    ▪ alternatively, use the .plot() method for Pandas Series

```python
# prepare data
# we wanna know how many organizations and privates do we have:

d = sales['Account'].value_counts()
# plot

plt.bar(d.index, d, width=.5)
plt.xlim(-.5, 1.5); plt.show()    # broaded xlim to make plot look nicer
```

##### Bar plot with pd.Series.plot( )

```python
# same plot using Pandas Series .plot() method:

sales['Account'].value_counts().plot(kind='bar', rot=0)   # rot=0 to suppress rotation by 90 deg. 
plt.show();
```

```python
from pandas.core.reshape.pivot import pivot_table
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Assumption: sales data already downloaded under name 'sales', se function defined

# recode sales variable: 


sales.loc[sales['Sale']==0, 'Sale']='no'
sales.loc[sales['Sale']==1, 'Sale']='yes'
sales #✅ 

# create data frame with values to plot: 

df = sales.pivot_table('Value', index='Sale', columns='Account', aggfunc=['mean', se]).stack()
#define width of bars
barwidth = .25

#prepare data
groups = np.unique(sales['Sale'])
x1 = np.arange(len(groups))
x2 = x1 + barwidth 
y1 = df.loc[groups[0]]['mean']
se1 = df.loc[groups[0]]['se']
y2 = df.loc[groups[1]]['mean']
se2 = df.loc[groups[1]]['se']


# plot bars: 

plt.bar(x1, y1, yerr=se1, width=barwidth, edgecolor='white', label=groups[0])
plt.bar(x2, y2, yerr=se1, width=barwidth, edgecolor='white', label=groups[1])

#fine tuning (ince ayar): add xticks on the middile of the group bars, add axes labels and title

plt.xticks(x1, + 5*barwidth, np.unique(sales['Account']))
plt.xlabel('Account type'); plt.ylabel('Mean order value')
plt.title('A grouped bar plot with error bars')

# create legend, show graphic 

plt.legend(title='Sale', loc='best'); plt.show()
```
